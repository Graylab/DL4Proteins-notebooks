### <b> <font color='#569098'> Install necessary software for local use

* Setup VSCode: https://code.visualstudio.com/docs/setup/setup-overview
* Install conda: https://docs.conda.io/projects/conda/en/stable/user-guide/getting-started.html
* Setup Jupyter use for VSCode: https://www.youtube.com/watch?v=h1sAzPojKMg

### <b> <font color='#569098'> Download the local DL4Proteins files

* Download all the content in the local notebooks folder
```
DL4Proteins-notebooks/local/*
```
* To use a given notebook locally, create a conda environment using the associated yml file. For example, to use notebook 1, create a conda env for notebook 1 using the workshop1 yml file
```
conda env create -f workshop1.yml
```
* Open up your Jupyter notebook and for your python interpreter select the your workshop 1 environment
